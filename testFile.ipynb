```python
strPath = r'D:\边缘点.txt'
pointIn = []
with open(strPath) as f:
    for tempLine in f:
        tempPoint = list(map(float,tempLine.split(',')[0:3]))
        pointIn.append(tempPoint)
print(pointIn)
```

```python
tempPointIn = list(zip(*pointIn))
print(tempPointIn)
```

```python
dXMax = max(tempPointIn[1])
dXMin = min(tempPointIn[1])
dYMax = max(tempPointIn[2])
dYMin = min(tempPointIn[2])
print(dXMax,dXMin,dYMax,dYMin)
```
